1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

### Accuracy
Scenario: Quality control in manufacturing, where correctly identifying both defective and non-defective products is important.
Rationale: In scenarios where both positive and negative outcomes have a similar impact, accuracy provides a straightforward assessment of overall model performance.

### Sensitivity
Scenario: Medical screening for diseases (e.g., cancer detection).
Rationale: In these situations, it is critical to identify as many true positive cases as possible to ensure early diagnosis and treatment, even if it means allowing for some false positives.

### Specificity
Scenario: Spam email filters, where it's important to correctly identify non-spam (legitimate) emails.
Rationale: High specificity ensures that legitimate emails are not wrongly classified as spam, minimizing inconvenience to users.

### Precision
Scenario: Fraud detection in banking transactions.
Rationale: Precision is important when the cost of false positives is high, as incorrectly flagging legitimate transactions as fraudulent can inconvenience customers and reduce trust in the system.

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

ab_reduced = ab.drop(['Weight_oz', 'Width', 'Height'], axis=1)

ab_reduced_noNaN = ab_reduced.dropna()

ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

print("Initial DataFrame Overview:\n")
print(ab_reduced_noNaN.info())
print("\nSummary Statistics:\n")
print(ab_reduced_noNaN.describe())
print("\nCategory Distribution:\n")
print(ab_reduced_noNaN['Hard_or_Paper'].value_counts())

ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

print(f"\nNumber of observations in training set: {len(ab_reduced_noNaN_train)}")
print(f"Number of observations in testing set: {len(ab_reduced_noNaN_test)}")

X = ab_reduced_noNaN_train[['List Price']]
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']

clf = tree.DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paper', 'Hard'], filled=True)
plt.show()
````

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, recall_score, precision_score
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

ab_reduced = ab.drop(['Weight_oz', 'Width', 'Height'], axis=1)

ab_reduced_noNaN = ab_reduced.dropna()

ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

print("Initial DataFrame Overview:\n")
print(ab_reduced_noNaN.info())
print("\nSummary Statistics:\n")
print(ab_reduced_noNaN.describe())
print("\nCategory Distribution:\n")
print(ab_reduced_noNaN['Hard_or_Paper'].value_counts())

ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

print(f"\nNumber of observations in training set: {len(ab_reduced_noNaN_train)}")
print(f"Number of observations in testing set: {len(ab_reduced_noNaN_test)}")

X_train = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']

X_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

clf2 = tree.DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train, y_train)

plt.figure(figsize=(20, 10))
tree.plot_tree(clf2, feature_names=['NumPages', 'Thick', 'List Price'], class_names=['Paper', 'Hard'], filled=True)
plt.show()

y_pred_clf2 = clf2.predict(X_test)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)
ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=['Paper', 'Hard']).plot()
plt.show()

accuracy_clf2 = accuracy_score(y_test, y_pred_clf2)
sensitivity_clf2 = recall_score(y_test, y_pred_clf2)
specificity_clf2 = cm_clf2[0, 0] / (cm_clf2[0, 0] + cm_clf2[0, 1])

print(f"Accuracy for clf2: {accuracy_clf2:.2f}")
print(f"Sensitivity for clf2: {sensitivity_clf2:.2f}")
print(f"Specificity for clf2: {specificity_clf2:.2f}")

X_train_clf = ab_reduced_noNaN_train[['List Price']]
X_test_clf = ab_reduced_noNaN_test[['List Price']]

clf = tree.DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train_clf, y_train)

y_pred_clf = clf.predict(X_test_clf)
cm_clf = confusion_matrix(y_test, y_pred_clf)
ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=['Paper', 'Hard']).plot()
plt.show()

accuracy_clf = accuracy_score(y_test, y_pred_clf)
sensitivity_clf = recall_score(y_test, y_pred_clf)
specificity_clf = cm_clf[0, 0] / (cm_clf[0, 0] + cm_clf[0, 1])

print(f"Accuracy for clf: {accuracy_clf:.2f}")
print(f"Sensitivity for clf: {sensitivity_clf:.2f}")
print(f"Specificity for clf: {specificity_clf:.2f}")
````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions.
````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, recall_score, precision_score
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

ab_reduced = ab.drop(['Weight_oz', 'Width', 'Height'], axis=1)

ab_reduced_noNaN = ab_reduced.dropna()

ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

print("Initial DataFrame Overview:\n")
print(ab_reduced_noNaN.info())
print("\nSummary Statistics:\n")
print(ab_reduced_noNaN.describe())
print("\nCategory Distribution:\n")
print(ab_reduced_noNaN['Hard_or_Paper'].value_counts())

ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

print(f"\nNumber of observations in training set: {len(ab_reduced_noNaN_train)}")
print(f"Number of observations in testing set: {len(ab_reduced_noNaN_test)}")

X_train = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']

X_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

clf2 = tree.DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train, y_train)

plt.figure(figsize=(20, 10))
tree.plot_tree(clf2, feature_names=['NumPages', 'Thick', 'List Price'], class_names=['Paper', 'Hard'], filled=True)
plt.show()

importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

print("\nFeature Importances:")
for name, importance in zip(feature_names, importances):
    print(f"{name}: {importance:.4f}")

most_important_feature = feature_names[np.argmax(importances)]
print(f"\nMost important predictor variable: {most_important_feature}")

y_pred_clf2 = clf2.predict(X_test)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)
ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=['Paper', 'Hard']).plot()
plt.show()

accuracy_clf2 = accuracy_score(y_test, y_pred_clf2)
sensitivity_clf2 = recall_score(y_test, y_pred_clf2)
specificity_clf2 = cm_clf2[0, 0] / (cm_clf2[0, 0] + cm_clf2[0, 1])

print(f"Accuracy for clf2: {accuracy_clf2:.2f}")
print(f"Sensitivity for clf2: {sensitivity_clf2:.2f}")
print(f"Specificity for clf2: {specificity_clf2:.2f}")

X_train_clf = ab_reduced_noNaN_train[['List Price']]
X_test_clf = ab_reduced_noNaN_test[['List Price']]

clf = tree.DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train_clf, y_train)

y_pred_clf = clf.predict(X_test_clf)
cm_clf = confusion_matrix(y_test, y_pred_clf)
ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=['Paper', 'Hard']).plot()
plt.show()

accuracy_clf = accuracy_score(y_test, y_pred_clf)
sensitivity_clf = recall_score(y_test, y_pred_clf)
specificity_clf = cm_clf[0, 0] / (cm_clf[0, 0] + cm_clf[0, 1])

print(f"Accuracy for clf: {accuracy_clf:.2f}")
print(f"Sensitivity for clf: {sensitivity_clf:.2f}")
print(f"Specificity for clf: {specificity_clf:.2f}")
````

In linear regression, coefficients represent the magnitude and direction of the relationship between each predictor and the outcome variable, with each coefficient showing how much the dependent variable changes for a one-unit change in the corresponding independent variable, holding all other variables constant. In decision trees, feature importance is a measure of how much a feature contributes to the decision-making process, based on its ability to reduce impurity (like Gini or entropy) at each split; higher importance indicates a stronger influence on the model's predictions. While linear regression focuses on the specific effect of predictors, decision trees capture nonlinear relationships and interactions between features.
